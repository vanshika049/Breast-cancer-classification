```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset
dataset = pd.read_csv('/kaggle/input/breastcancer-dataset/breast_cancer_data.csv')
```

```python
dataset.head()
```

```python
print(dataset)
```

```python
# print last 5 rows of the dataframe
dataset.tail()
```

```python
# number of rows and columns in the dataset
dataset.shape
```

```python
# getting some information about the data
dataset.info()
```

```python
dataset['diagnosis'] = dataset['diagnosis'].map({'M':'0',
                                                 'B':'1',})
```

```python
dataset.drop(columns=['id','Unnamed: 32'],inplace=True)
dataset.info()
```

```python
dataset.describe()
```

```python
# checking the distribution of Target Varibale
dataset['diagnosis'].value_counts()
```

```python
dataset.groupby('diagnosis').mean()
```

# Separating the features and target

```python
X = dataset.drop(columns='diagnosis', axis=1)
Y = dataset['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

# Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
```

# Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```
